```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4]
y = [39.03, 38.49, 38.44, 38.29]
```

```python
plt.plot(x,y,'g-',label='MSVD-QA:8')
#plt.legend('MSVD-QA:8',loc="upper left")
#plt.text(0.5,0.5,'Accuracy')
plt.xlabel('Iteration steps',fontsize='large')
plt.ylabel('Accuracy',fontsize='large')
plt.grid(alpha=0.3)
plt.ylim(38.2,39.2)
for a, b in zip(x,y):
    plt.text(a, b+0.001,'%.2f' % b, ha='center', va= 'bottom',fontsize=12)
plt.legend(fontsize='large')
plt.savefig('MSVD_steps_new.png')
```

```python
# SVQA
x = [4,8,12,16,20,24]
y = [49.45, 49.09, 49.07, 50.29, 50.38, 49.51]
import numpy as np
```

```python

plt.plot(x,y,'red',label='SVQA')
#plt.legend('MSVD-QA:8',loc="upper left")
#plt.text(0.5,0.5,'Accuracy')
plt.xlabel('Number of divided clips',fontsize='large')
plt.ylabel('Accuracy',fontsize='large')
plt.grid(alpha=0.3)
plt.ylim(48.5,51.0)
plt.xlim(2,26)
plt.xticks(np.arange(4,28,4))
for a, b in zip(x,y):
    plt.text(a, b+0.001,'%.2f' % b, ha='center', va= 'bottom',fontsize=12)
plt.legend(fontsize='large')
plt.savefig('SVQA_clips_new.png')
```
